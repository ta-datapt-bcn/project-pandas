## Functions

```python
def pshape(df):
    '''
        imprimeix el 'shape' del df passat i el de les dades originals
    '''
    sh1 = df.shape
    sh2 = data.shape
    print(f'df:{sh1} | data:{sh2}')
    
    
def printr(df,list_cols,num_rows=54):
    '''
        imprimeix alguns valors de les columnes passades per poder-les-hi comparar
    '''
    print ('LEN df = ',len(df))
    for index, row in df.iterrows():
        for col in list_cols:
            print (row[col])
        print('----------------')
        if index >= num_rows -1: break
```

### One Null is filled with href

```python
data['href formula'] = [y if len(str(x)) <4 else x for x,y in zip(data['href formula'],data['href']) ]
```

## 2- Duplicates

```python

columns_names = ['pdf','href formula','href']

yc = data[(data['href formula']==data['href'])]
nc = data[(data['href formula']!=data['href']) & (data['href'].isnull()==False)]
nc = nc.reset_index(drop=True)

# Print comparació dels df iguals i diferents
printr(yc,columns_names,2)
printr(nc,columns_names,2)

# es comprova que no queda cap NULL
naf = data[(data['href formula'].isnull())]
pshape(naf)

# es comprova que tots els pdf corresponen a 'href formula'

test = [x for x,y in zip(data['href formula'],data['pdf']) if y not in x  ]
print(test)

# per tant, s'inclouen les columnes 'pdf' i 'href' per eliminar, ja que són duplicats
```

```python
yc = data[(data['Case Number.1']==data['Case Number.2'])]
nc = data[(data['Case Number.1']!=data['Case Number.2'])]

na1 = data[(data['Case Number.1'].isnull()==True)]
na2 = data[(data['Case Number.2'].isnull()==True)]

pshape(yc)
pshape(nc)
pshape(na1)
pshape(na2)
```

## 3- Drop Columns

```python
# els números de cas 1 i 2 són alies identificatius, que ja tenim al 'Case Number', així que s'eliminaran

# també 'Year' és un duplicat de la data, es pot eliminar

# 'Date' està inclós a 'Case Number', es pot eliminar 

# 'original order' no aporta informació útil, s'elimina


drop_cols = ['Date','Year','pdf','href','Case Number.1','Case Number.2','Unnamed: 22','Unnamed: 23','original order']

data.drop(drop_cols, axis=1, inplace=True)
```
